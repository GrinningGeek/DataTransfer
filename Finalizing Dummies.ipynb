```python
import pandas as pd
import numpy as np
from pprint import pprint

df = pd.read_csv('Speeding/Violations_2017_Speeding_after_dummies.csv', parse_dates=['time_of_stop'])
```

```python
df.head()
```

```python
df.columns
```

```python
#Combining columns down into binary, remove one of the car makes
df = df.drop(['belts_No', 'personal_injury_No', 'property_damage_No', 'fatal_No', 
              'commercial_license_No', 'alcohol_No', 'work_zone_No', 'contributed_to_accident_No',
             'violation_type_Warning'], axis=1)
```

```python
df = df.rename(columns={"belts_Yes": "belts", "personal_injury_Yes" : "personal_injury", 
                        "property_damage_Yes" : "property_damage", "fatal_Yes" : "fatal", 
                        "commercial_license_Yes" : "commercial_license", "alcohol_Yes" : "alcohol", 
                        "work_zone_Yes" : "work_zone", 
                        "contributed_to_accident_Yes" : "contributed_to_accident",
                        "violation_type_Citation" : "citation"})
```

```python
dataset = df.drop(['time_of_stop', 'latitude', 'longitude'], axis=1)
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
from sklearn import tree

X = dataset.drop(['citation'], axis=1)
Y = dataset.citation
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,Y)
```

```python
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render('citations')
```

```python
clf.score(X=X, y=Y)
```

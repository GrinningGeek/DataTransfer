```python
import pandas as pd
from arcgis.gis import GIS
from arcgis.geocoding import get_geocoders, batch_geocode
```

```python
gis = GIS("https://nga.maps.arcgis.com", "jonesjm_nga")
```

```python
df = pd.read_csv('Speeding/Violations_2017_Speeding_geoloc.csv')
```

```python
df = df.drop(['Geolocation'], axis=1)
blanks = df[df['Latitude'].isnull()]
```

```python
#blanks = blanks.reset_index(drop=True)
blanks.head()
```

```python
addresses = blanks['Location'].tolist()

results = batch_geocode(addresses)
```

```python
#convert list to dict
results_dict = dict()

geo_lat=list()
geo_lon=list()

for i in range(0, len(results)):
    #Convert each item to dict before grabbing the key values
    results_dict.update(results[i])
    
    geo_lat.append(results_dict.get('location')['y'])
    geo_lon.append(results_dict.get('location')['x'])
```

```python
#blanks['Latitude'] = geo_lat
#blanks['Longitude'] = geo_lon

blanks['Latitude'] = pd.Series(geo_lat).values
blanks['Longitude'] = pd.Series(geo_lon).values
```

```python
blanks
```

```python
#Trim unreal results based on geoloc

blanks = blanks[blanks["Latitude"] >= 38]
blanks = blanks[blanks["Latitude"] <= 40]
blanks = blanks[blanks["Longitude"] >= -78]
blanks = blanks[blanks["Longitude"] <= -75]
```

```python
import numpy as np
df = df[np.isfinite(df['Latitude'])]
```

```python
df.append(blanks, ignore_index=True)
df = df.reset_index(drop=True)
df.head()
```

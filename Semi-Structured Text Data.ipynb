```python
import pandas as pd
df = pd.read_csv('Speeding/Violations_2017_Speeding_after_geoloc.csv', parse_dates=[['Date Of Stop', 'Time Of Stop']])
```

```python
from difflib import SequenceMatcher
#text = """there are
#some toyot in my text
#but I cannot find them"""
#df = df[:100]
df['Make2'] = 'Make'
```

```python
df.head()
```

```python
def fuzzy_search(search_key, text, strictness):
    similarity = SequenceMatcher(None, text, search_key)
    if similarity.ratio() > strictness:
        return search_key

make_list = ['ACURA', 'ASTON MARTIN', 'AUDI', 'BMW', 'BENTLEY', 'BUICK',
'CADILLAC', 'CHEVROLET', 'CHRYSLER', 'DODGE', 'DUCATI', 'FERARRI',
'FIAT', 'FORD', 'FREIGHTLINER', 'GEO', 'GMC', 'GILLIG', 'GRUMMAN',
'HARLEY', 'HINO', 'HONDA', 'INFINITI', 'ISUZU', 'JAGUAR', 'HUMMER',
'HYUNDAI', 'INTERNATIONAL', 'JEEP', 'KAWASAKI', 'KENWORTH',
'KIA', 'LANDROVER', 'LEXUS', 'LINCOLN', 'MACK', 'MASERATI',
'MAZDA', 'MERCEDES', 'MERCURY', 'MINI', 'MITSUBISHI', 'NISSAN',
'OLDSMOBILE', 'PETERBILT', 'PLYMOUTH', 'POLARIS',
'PONTIAC', 'PORSCHE', 'RANGE ROVER', 'SATURN', 'SCION',
'SAAB', 'SMART', 'SUBARU', 'SUZUKI', 'TESLA', 'TOYOTA',
'VOLKSWAGEN', 'VOLVO', 'YAMAHA']

not_found = 0

for index, text in df.iterrows():
    found = 0
    #Special fix for Chevrolet...
    
    if ((text['Make'] == 'CHEVY') or (text['Make']=='CHEV')):
        text['Make'] = 'CHEVROLET'
    
    #Fix for Volkswagen...
    elif ((text['Make'] == 'VW') or (text['Make'] == 'VOLKS') or (text['Make'] == 'VOLK')):
        text['Make'] = 'VOLKSWAGEN'
    
    #fix for Oldsmobile...
    elif text['Make'] == 'OLDS':
        text['Make'] = 'OLDSMOBILE'
        
    #RAM is a type of Dodge...
    elif text['Make'] == 'RAM':
        text['Make'] = 'DODGE'
    
    #MERZ is abbreviation for Mercedes...
    elif ((text['Make'] == 'MERZ') or text['Make'] == 'MERC'):
        text['Make'] = 'MERCEDES'
        
    #MITZ and MITS is Mitsubishi...
    elif ((text['Make'] == 'MITS') or (text['Make'] == 'MITZ') or (text['Make'] == 'MITSU')):
        text['Make'] = 'MITSUBISHI'
        
    #HD is Harley
    elif text['Make'] == 'HD' or text['Make'] == 'H D':
        text['Make'] = 'HARLEY'
        
    #INTL is international:
    elif text['Make'] == 'INTL':
        text['Make'] = 'INTERNATIONAL'
        
        
    #CADI is cadillac:
    elif ((text['Make'] == 'CADI') or (text['Make'] == 'CAD')):
        text['Make'] = 'CADILLAC'
    
    #CHRY is chrysler
    elif text['Make'] == 'CHRY':
        text['Make'] = 'CHRYSLER'
        
    #HON is honda
    elif text['Make'] == 'HON':
        text['Make'] = 'HONDA'
        
    #HYUN is hyundai
    elif text['Make'] == 'HYUN':
        text['Make'] = 'HYUNDAI'
        
    #infi is infiniti
    elif text['Make'] == 'INFI':
        text['Make'] = 'INFINITI'
        
    elif text['Make'] == 'LINC':
        text['Make'] = 'LINCOLN'
    
    elif text['Make'] == 'LNDR':
        text['Make'] = 'LANDROVER'
        
    elif text['Make'] == 'MAZ':
        text['Make'] = 'MAZDA'
    
    elif text['Make'] == 'KAWK':
        text['Make'] = 'KAWASAKI'
        
    elif text['Make'] == 'LEX':
        text['Make'] = 'LEXUS'
    
    elif text['Make'] == 'BENT':
        text['Make'] = 'BENTLEY'
        
    elif text['Make'] == 'PONT':
        text['Make'] = 'PONTIAC'
        
    elif text['Make'] == 'PORS':
        text['Make'] = 'PORSCHE'
        
    elif text['Make'] == 'SUB':
        text['Make'] = 'SUBARU'
    
    elif text['Make'] == 'TOY':
        text['Make'] = 'TOYOTA'
        
    elif ((text['Make'] == 'RANG') or (text['Make'] == 'RANGE')):
        text['Make'] = 'RANGE ROVER'
    
    elif text['Make'] == 'YAM':
        text['Make'] = 'YAMAHA'
    
    elif text['Make'] == 'PTRB':
        text['Make'] = 'PETERBILT'
        
    elif text['Make'] == 'NISS':
        text['Make'] = 'NISSAN'
        
    elif text['Make'] == 'MNNI':
        text['Make'] = 'MINI'
        
    elif text['Make'] == 'JAG':
        text['Make'] = 'JAGUAR'
        
    elif text['Make'] == 'KW':
        text['Make'] = 'KENWORTH'
        
    elif text['Make'] == 'ISU':
        text['Make'] = 'ISUZU'
        
    elif text['Make'] == 'FRHT':
        text['Make'] = 'FREIGHTLINER'
        
    elif text['Make'] == 'PLYM':
        text['Make'] = 'PLYMOUTH'
    
    #print(text)
    for item in make_list:
        if ((fuzzy_search(item, text['Make'], 0.75) != None) and found == 0):
            #print (fuzzy_search(item, text['Make'], 0.65))
            df.at[index, 'Make2'] = (fuzzy_search(item, text['Make'], 0.75))
            found = 1
            
            
    if found == 0:
        #print(text)
        not_found += 1
        df.at[index, 'Make2'] = 'Unknown'
    
```

## Categorical Data

```python
df1 = df[['Date Of Stop_Time Of Stop', 'Latitude', 'Longitude',
        'Belts', 'Personal Injury', 'Property Damage', 'Fatal',
        'Commercial License', 'Alcohol', 'Work Zone', 
          'Contributed To Accident', 'Violation Type', 'Make', 'Make2']]

df1.columns = ['time_of_stop', 'latitude', 'longitude', 
               'belts', 'personal_injury', 'property_damage', 'fatal', 
               'commercial_license', 'alcohol', 'work_zone', 
               'contributed_to_accident', 'violation_type', 'make', 'make2']
```

```python
df1.head()
```

```python
#Scrub that which cannot be fixed
#No blank lats/longs/make2
df_scrub = df1[(df1.latitude.notnull()) & \
        (df1.longitude.notnull()) & \
        (df1.make2 != 'Unknown')]

print(str(len(df1) - len(df_scrub)) + ' items removed.')

df = df_scrub
```

```python
#Fix for the rovers
df.loc[df['make2'] == 'LANDROVER', 'make2'] = 'LAND_ROVER'
df.loc[df['make2'] == 'RANGE ROVER', 'make2'] = 'RANGE_ROVER'
```

```python
#Drop extra make column, rename
df.drop(['make'], axis=1, inplace=True)
df = df.rename(columns={"make2": "make"})
```

```python
df.to_csv('Speeding/Violations_2017_Speeding_before_dummies.csv', index=False)
```

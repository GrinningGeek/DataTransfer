```python
import pandas as pd

mydateparser = lambda x:pd.datetime.strptime(x, "%m/%d/%Y")

df = pd.read_csv('Traffic_Violations.csv', 
                 parse_dates=['Date Of Stop'],
                 date_parser=mydateparser)

df['Stop Year'] = df['Date Of Stop'].map(lambda x: x.year)

df_2017 = df.loc[df['Stop Year']==2017]

df_2017 = df_2017.drop(['Stop Year'], axis=1)
```

```python
df_2012 = df.loc[df['Stop Year']==2012]
df_2012 = df_2012.drop(['Stop Year'], axis=1)

df_2013 = df.loc[df['Stop Year']==2013]
df_2013 = df_2013.drop(['Stop Year'], axis=1)

df_2014 = df.loc[df['Stop Year']==2014]
df_2014 = df_2014.drop(['Stop Year'], axis=1)

df_2015 = df.loc[df['Stop Year']==2015]
df_2015 = df_2015.drop(['Stop Year'], axis=1)

df_2016 = df.loc[df['Stop Year']==2016]
df_2016 = df_2016.drop(['Stop Year'], axis=1)

df_2012 = df.loc[df['Stop Year']==2012]
df_2012 = df_2012.drop(['Stop Year'], axis=1)

df_2012.to_csv('All/Violations_2012.csv', index=False)
df_2013.to_csv('All/Violations_2013.csv', index=False)
df_2014.to_csv('All/Violations_2014.csv', index=False)
df_2015.to_csv('All/Violations_2015.csv', index=False)
df_2016.to_csv('All/Violations_2016.csv', index=False)
df_2017.to_csv('All/Violations_2017.csv', index=False)
```

```python
df_2012_influence = df_2012[df_2012['Description'].str.contains("INFLUENCE", 
                                                                na=False)]

df_2013_influence = df_2013[df_2013['Description'].str.contains("INFLUENCE", 
                                                                na=False)]

df_2014_influence = df_2014[df_2014['Description'].str.contains("INFLUENCE", 
                                                                na=False)]

df_2015_influence = df_2015[df_2015['Description'].str.contains("INFLUENCE", 
                                                                na=False)]

df_2016_influence = df_2016[df_2016['Description'].str.contains("INFLUENCE", 
                                                                na=False)]

df_2017_influence = df_2017[df_2017['Description'].str.contains("INFLUENCE", 
                                                                na=False)]

df_2012_influence.to_csv('Influence/Violations_2012_Influence.csv', index=False)
df_2013_influence.to_csv('Influence/Violations_2013_Influence.csv', index=False)
df_2014_influence.to_csv('Influence/Violations_2014_Influence.csv', index=False)
df_2015_influence.to_csv('Influence/Violations_2015_Influence.csv', index=False)
df_2016_influence.to_csv('Influence/Violations_2016_Influence.csv', index=False)
df_2017_influence.to_csv('Influence/Violations_2017_Influence.csv', index=False)
```
